```
import numpy as np
H = 2.3 # metri a pieno carico
D = 2.1 # metri 
V = H * np.pi * (D/2)**2
print ('Pieno: {:.0f} litri'.format(V*1000))
A0 = 1.63 # primo anelllo coperto
V = A0 * np.pi * (D/2)**2
print ('Primo anello coperto: {:.0f} litri'.format(V*1000))
A = 0.257 # ogni anello e' 20 cm
VA = A * np.pi * (D/2)**2

for i in range (0, 5):
  V = V - VA
  print ('anello {:d} coperto: {:.0f} litri'.format(i+1,V*1000))

A5 = 0.345
V5 = A5 * np.pi * (D/2)**2
print ('Ultimo anello coperto: {:.0f} litri (calcolato dalla misura del fondo)'.format(V5*1000))
```
