```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fast')
plt.rc('font', size = 14, family ='Sans-Serif')
# reference https://arxiv.org/abs/1602.03781
#s0 = 10**-42  # DM-nucleus scattering cross section in the zero momentum transfer limit
F = 1.0       # F(Er) is the nuclear form factor of the target nucleus
rho_x = 0.3   # GeVlocal DM density and v is the relative velocity 
v = 300*1e3   # m / s
c = 299792458 # m / s
# see also http://www.hep.ph.ic.ac.uk/susytalks/fairbairnimperialoneday.pdf
```

```
Mx = 6.0      # DM Mas GeV (1.79 x 10-27)
Mb = 4.0      # Nucleai Mass GeV
mu = Mx*Mb/(Mx+Mb)
# nuclear recoil per He puro
theta = np.linspace(0,np.pi,20)
#
E_r = 2*(v**2/c**2) * (mu**2/Mb) * np.cos(theta)**2 # eq 5 https://arxiv.org/abs/1602.03781 
#
plt.plot(theta, E_r*10**6, label='$M_\chi$ = {0:.1f} GeV $v$ = {1:.1f} km/s'.format(Mx, v/1e3))
plt.xlabel('Theta')
plt.ylabel('Recoil energy [keV]')
plt.legend()
plt.title('He')
plt.show()
```

```
Mx = 6.0         # DM Mas GeV (1.79 x 10-27)
MHeCF4 = 4.0*0.6+88*0.4        # Nucleai Mass GeV
mu_HeCF4 = Mx*MHeCF4/(Mx+MHeCF4)
# nuclear recoil per HeCF4
theta = np.linspace(0,np.pi,20)
#
E_r = 2*(v**2/c**2) * (mu_HeCF4**2/MHeCF4) * np.cos(theta)**2 # eq 5 https://arxiv.org/abs/1602.03781 
#
plt.plot(theta, E_r*10**6, label='$M_\chi$ = {0:.1f} GeV $v$ = {1:.1f} km/s'.format(Mx, v/1e3))
plt.title('HeCF4 60/40')
plt.xlabel('Theta')
plt.ylabel('Recoil energy [keV]')
plt.legend()
plt.show()
```

```
MXe = 131.2      # Nucleai Mass GeV
mu_Xe = Mx*MXe/(Mx+MXe)
sen_Xe_HeCF4 = (mu_Xe**2/MXe)*(MHeCF4/mu_HeCF4**2) # eq 20 https://arxiv.org/abs/1602.03781 
print('Sensibilita equivalente in massa: {:.2f} a Mx = {:.1f} GeV'.format(sen_Xe_HeCF4, Mx))
```
