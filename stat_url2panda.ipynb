```
# open XML URL to Pand
# fonte Open data http://www.datiopen.it/ sito di aggregazione --> Scarica --> ottine URL 
def xml2df(xml_data):
    import xml.etree.ElementTree as ET
    import pandas as pd
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
        all_records.append(record)
    df = pd.DataFrame(all_records)
    return df

import requests
url_xml_file ='http://www.user-agents.org/allagents.xml'
xml_data = requests.get(url_xml_file).content
#xml_data = open('file.xml').read()
df = xml2df(xml_data)
df.tail()
```
